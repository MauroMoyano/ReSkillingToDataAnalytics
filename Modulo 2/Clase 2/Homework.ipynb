```python
import csv
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}

archivo = open('Emisiones_CO2.csv', 'r')

lector = csv.reader(archivo)

for fila in lector:
    type(fila)
    partes = fila[0].split('|')
    #print(len(partes))
    #print(partes)
    try:
        dicc_emisiones['cod_pais'].append(partes[0])
        dicc_emisiones['nom_pais'].append(partes[1])
        dicc_emisiones['region'].append(partes[2])
        dicc_emisiones['anio'].append(partes[3])
        dicc_emisiones['co2'].append(partes[4])
        dicc_emisiones['co2_percapita'].append(partes[5])
    except IndexError:
        print("rompe")
print(dicc_emisiones)
    
```

```python
for clave in dicc_emisiones:
    while '' in dicc_emisiones[clave]:
        if '' in dicc_emisiones[clave]:
            dicc_emisiones[clave].remove('')
```
